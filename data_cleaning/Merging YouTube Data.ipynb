```python
import numpy as np
import pandas as pd
import json
```

```python
# Load original data set
ted_data = pd.read_csv('../src/data/ted_main.csv')
ted_data
```

```python
# Generate names of JSON files
json_files = []
for i in np.arange(0, 2550, 10):
    json_files.append("./youtube_data/video_data_" + str(i) + "_" + str(i+10) + ".json")
```

```python
# Load json files
json_data = []
for filepath in json_files:
    datalist = json.load(open(filepath))
    for data in datalist:
        json_data.append(data)
```

```python
# Initialize columns
ted_data["youtube_title"] = np.NaN
ted_data["youtube_description"] = np.NaN
ted_data["youtube_videoId"] = np.NaN
ted_data["youtube_videoLink"] = np.NaN
ted_data["youtube_publishedAt"] = np.NaN
ted_data["youtube_viewCount"] = np.NaN
ted_data["youtube_likeCount"] = np.NaN
ted_data["youtube_dislikeCount"] = np.NaN
ted_data["youtube_favoriteCount"] = np.NaN
ted_data["youtube_commentCount"] = np.NaN
```

```python
for i in range(len(json_data)):
    print("Reading json block ", i)
    ted_data.loc[i, "youtube_title"] = json_data[i]['items'][0]['snippet'].get('title')
    ted_data.loc[i, "youtube_description"] = json_data[i]['items'][0]['snippet'].get('description')
    ted_data.loc[i, "youtube_videoId"] = json_data[i]['items'][0]['id'].get('videoId')
    ted_data.loc[i, "youtube_videoLink"] = "https://www.youtube.com/watch?v=" + json_data[i]['items'][0]['id'].get('videoId')
    ted_data.loc[i, "youtube_publishedAt"] = json_data[i]['items'][0]['snippet'].get('publishedAt')
    ted_data.loc[i, "youtube_viewCount"] = json_data[i]['items'][0]['stats']['items'][0]['statistics'].get('viewCount')
    ted_data.loc[i, "youtube_likeCount"] = json_data[i]['items'][0]['stats']['items'][0]['statistics'].get('likeCount')
    ted_data.loc[i, "youtube_dislikeCount"] = json_data[i]['items'][0]['stats']['items'][0]['statistics'].get('dislikeCount')
    ted_data.loc[i, "youtube_favoriteCount"] = json_data[i]['items'][0]['stats']['items'][0]['statistics'].get('favoriteCount')
    ted_data.loc[i, "youtube_commentCount"] = json_data[i]['items'][0]['stats']['items'][0]['statistics'].get('commentCount')

ted_data.to_csv("ted_merged_yt.csv", index=False)
print("Finished!")
```
