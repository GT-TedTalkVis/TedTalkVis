```python
import numpy as np
import pandas as pd
```

```python
ted_data = pd.read_csv('./ted_merged_yt.csv')
ted_data
```

```python
def textSimilarity(str1, str2):
    # Create vectors representing the string
    v1 = np.zeros((16384), dtype=int)
    v2 = np.zeros((16384), dtype=int)
    
    for i in range(len(str1)):
        code = ord(str1[i])
        if code < 16384 and code > 32:
            v1[code] += 1
    
    for i in range(len(str2)):
        code = ord(str2[i])
        if code < 16384 and code > 32:
            v2[code] += 1
    
    # Calculate similarity
    return np.dot(v1 / np.linalg.norm(v1), v2 / np.linalg.norm(v2))
```

```python
# Check if the TED site titles and the YouTube titles are similar
similarity_boolArray = []
for i in range(len(ted_data)):
    similarity = textSimilarity(ted_data.loc[i, "name"], ted_data.loc[i, "youtube_title"])
    similarity_boolArray.append(similarity < 0.9)
similarity_boolArray = np.array(similarity_boolArray)
```

```python
ted_data.loc[similarity_boolArray, ["name", "youtube_title"]]
```

```python
# Just add similarity score to csv
ted_data["name_yt_similarity"] = np.NaN
```

```python
for i in range(len(ted_data)):
    similarity = textSimilarity(ted_data.loc[i, "name"], ted_data.loc[i, "youtube_title"])
    ted_data.loc[i, "name_yt_similarity"] = similarity
```

```python
similarityCheck = ted_data.loc[:, ["name", "youtube_title", 'name_yt_similarity', "youtube_description", "youtube_videoId", 'url', 'youtube_videoLink', 'youtube_publishedAt',
       'youtube_viewCount', 'youtube_likeCount', 'youtube_dislikeCount',
       'youtube_favoriteCount', 'youtube_commentCount']]
```

```python
similarityCheck.to_csv("ted_youtube_similarity.csv")
```
